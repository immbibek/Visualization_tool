```python
# label title
plt.title('Rohit vs kohli')
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
# color and line widht

plt.title('Rohit vs kohli')
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='D')
plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
# legend --> location
plt.title('Rohit vs kohli')
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='D',label='Rohit')
plt.xlabel('season')
plt.ylabel('Runs scored')


plt.legend(loc='upper left') # if i wnt to run label i want to call legend()
```

```python
# limiting axes
price=[4500,6000,900,8000,100000] #due to  outliers
year=[2015,2016,2017,2018,2019]
plt.plot(year,price)
# to manage this
plt.ylim(0,8000)
```

```python
# grid
plt.title('Rohit vs kohli')
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='D',label='Rohit')
plt.xlabel('season')
plt.ylabel('Runs scored')
plt.grid()
```

```python
# show
plt.title('Rohit vs kohli')
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='D',label='Rohit')
plt.xlabel('season')
plt.ylabel('Runs scored')
plt.grid()

plt.show()
```
